```python
# pickle转csv 完整
import pickle    
import pandas as pd
test = pickle.load(open('/Users/zhongshannan/Documents/fakenews_detection/dataset/weiboA/validate.pickle', 'rb'))

for i in range(len(test['text'])):
    test['text'][i] = test['text'][i][:-1]
    #test['image'][i] = test['image'][i][0][:-4]  # 取到jpg
    test['image'][i] = test['image'][i][0] # 全取
    test['label'][i] = "real" if test['label'][i] == 0 else "fake" # label从01转换fake real
    # test['label'][i] = test['label'][i]
# print(len(test['text']),type(test['text']),len(test['image']),type(test['image']),len(test['label']),type(test['label']))

test1 = pd.DataFrame(columns = ['text'] , data=test['text'])
test2 = pd.DataFrame(columns = ['image_filenames'] , data=test['image'])
test3 = pd.DataFrame(columns = ['labels'] , data=test['label'])
df = pd.concat([test1,test2,test3], axis=1)
df.to_csv("/Users/zhongshannan/Documents/fakenews_detection/dataset/weiboA/weibo_validate.csv", encoding='utf-8', index=False)
```
