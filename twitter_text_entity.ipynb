```python
import pickle
train = pickle.load(open('/Users/zhongshannan/Documents/fakenews_detection/dataset/twitter/train_image_entity.pickle','rb'))
val = pickle.load(open('/Users/zhongshannan/Documents/fakenews_detection/dataset/twitter/val_image_entity.pickle','rb'))
test = pickle.load(open('/Users/zhongshannan/Documents/fakenews_detection/dataset/twitter/test_image_entity.pickle','rb'))
val.keys()
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification

tokenizer = AutoTokenizer.from_pretrained("Jean-Baptiste/roberta-large-ner-english")
model = AutoModelForTokenClassification.from_pretrained("Jean-Baptiste/roberta-large-ner-english")
```

```python
from transformers import pipeline
nlp = pipeline('ner', model=model, tokenizer=tokenizer, aggregation_strategy="simple")

ii = 1
val['text_entity'] = []

for text in val['text']:
    sen = ''
    entity = nlp(text)
    for i in range(len(entity)):
        sen += entity[i]['word'] + ' '
    val['text_entity'].append(sen[:-1])
    print(ii, sen[:-1])
    ii += 1

with open('/Users/zhongshannan/Documents/fakenews_detection/dataset/twitter/val_entity.pickle', 'wb') as f:
    pickle.dump(val, f)
```
